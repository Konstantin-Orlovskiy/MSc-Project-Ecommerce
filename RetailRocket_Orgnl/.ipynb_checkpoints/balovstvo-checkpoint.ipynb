```python
# Visualize all events
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig1 = plt.figure(figsize=(12, 12), dpi= 80, facecolor='w', edgecolor='k')
xyz = fig1.add_subplot(111, projection='3d')

x = df_events['timestamp'].astype(int)
y = df_events['visitorid'].astype(int)
z = df_events['itemid'].astype(int)
t = df_events['event'].astype(int)

xyz.set_xlabel('Timestamp')
xyz.set_ylabel('VisitorID')
xyz.set_zlabel('ItemID')
xyz.set_title('All Events')

markersize=0.1
img = xyz.scatter(x, y, z, c=t, cmap='Accent', s=markersize)

plt.show()
```

```python
# Let's apply PCA on the sparse matrix.

from scipy.sparse import csr_matrix
row = df_events['visitorid'].astype(int)
col = df_events['itemid'].astype(int)

df_tmp = df_events
#df_tmp['event'].replace(to_replace=dict(view=1, addtocart=2, transaction=3), inplace=True)

data = df_tmp['event'].astype(int)
all_visitors_items_grid = csr_matrix((data, (row, col)), shape=None)

```

```python
import matplotlib.pylab as plt
plt.spy(all_visitors_items_grid)
```

```python
# Let's apply PCA on the sparse matrix for transactions only.

from scipy.sparse import csr_matrix
row = df_transaction['visitorid'].astype(int)
col = df_transaction['itemid'].astype(int)

df_tmp = df_transaction
data = df_tmp['event'].astype(int)
transactions_visitors_items_grid = csr_matrix((data, (row, col)), shape=None)
```

```python
from sklearn.decomposition import TruncatedSVD

# Create Truncated Singular Value Decomposition
tsvd = TruncatedSVD(n_components=10)
```

```python
# Conduct TSVD on sparse matrix
Matrix_sparse_tsvd = tsvd.fit(all_visitors_items_grid).transform(all_visitors_items_grid)
```

```python
# Show results
print('Original number of features:', all_visitors_items_grid.shape[1])
print('Reduced number of features:', Matrix_sparse_tsvd.shape[1])
```
